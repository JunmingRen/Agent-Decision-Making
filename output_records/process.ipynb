```python
import re
from collections import defaultdict
import os

def extract_student_scores(content):
    """
    从文件内容中提取每个学生的评分数据
    """
    # 正则表达式匹配学生名字和评分数据
    pattern = r'Student (\w+):\s*((?:\d+\.\s*\d+\s*)+|(?:\d+\s*)+)'
    matches = re.findall(pattern, content)
    
    # 存储每个学生的评分数据
    student_scores = defaultdict(list)
    
    for student, scores in matches:
        # 处理有序号和无序号的格式
        if '.' in scores:
            # 有序号格式（如 "1. 5"），提取数字部分
            score_values = re.findall(r'\d+\.\s*(\d+)', scores)
        else:
            # 无序号格式（如 "5"），直接提取数字
            score_values = re.findall(r'\d+', scores)
        
        # 将评分数据转换为整数
        score_values = [int(value) for value in score_values]
        student_scores[student].append(score_values)
    
    return student_scores

def calculate_average(scores):
    """
    计算评分数据的平均值
    """
    if not scores:
        return 0
    return sum(scores) / len(scores)

def process_files(file_paths):
    """
    处理多个文件，提取每个学生的评分数据并计算平均值
    """
    results = {}
    
    for file_path in file_paths:
        with open(file_path, 'r') as file:
            content = file.read()
        
        # 提取学生评分数据
        student_scores = extract_student_scores(content)
        
        # 记录每个学生的第一次和第二次出现的数据
        file_results = {}
        for student, scores_list in student_scores.items():
            if len(scores_list) >= 2:
                # 记录第一次和第二次出现的数据
                first_scores = scores_list[0]
                second_scores = scores_list[1]
                
                # 计算平均值
                first_avg = calculate_average(first_scores)
                second_avg = calculate_average(second_scores)
                
                file_results[student] = {
                    'first_scores': first_scores,
                    'first_avg': first_avg,
                    'second_scores': second_scores,
                    'second_avg': second_avg,
                    'difference': abs(first_avg - second_avg)  # 计算差值
                }
        
        # 将文件结果保存到总结果中
        results[file_path] = file_results
    
    return results

def aggregate_results(results):
    """
    统计所有文档中每位学生的两次数据的平均数和差值
    """
    # 存储所有文档中每位学生的统计结果
    student_aggregates = defaultdict(lambda: {
        'first_avg_total': 0,
        'second_avg_total': 0,
        'difference_total': 0,
        'count': 0
    })
    
    # 遍历所有文件的结果
    for file_results in results.values():
        for student, data in file_results.items():
            # 累加第一次和第二次的平均值以及差值
            student_aggregates[student]['first_avg_total'] += data['first_avg']
            student_aggregates[student]['second_avg_total'] += data['second_avg']
            student_aggregates[student]['difference_total'] += data['difference']
            student_aggregates[student]['count'] += 1
    
    # 计算每位学生的平均值
    for student, aggregates in student_aggregates.items():
        count = aggregates['count']
        aggregates['first_avg_mean'] = aggregates['first_avg_total'] / count
        aggregates['second_avg_mean'] = aggregates['second_avg_total'] / count
        aggregates['difference_mean'] = aggregates['difference_total'] / count
    
    return student_aggregates

# 获取所有符合条件的文件路径
file_paths = []
for file in os.listdir():
    if file.startswith("consistency_test_20250111") and file.endswith(".txt"):
        file_paths.append(file)

# 处理文件并输出结果
results = process_files(file_paths)

# 统计所有文档中每位学生的两次数据的平均数和差值
student_aggregates = aggregate_results(results)

# 打印结果
print("所有文档中每位学生的统计结果：")
for student, aggregates in student_aggregates.items():
    print(f"学生: {student}")
    print(f"第一次评分的平均数的平均数: {aggregates['first_avg_mean']:.4f}")
    print(f"第二次评分的平均数的平均数: {aggregates['second_avg_mean']:.4f}")
    print(f"两次评分差值的平均数: {aggregates['difference_mean']:.4f}")
    print()
```

```python
import re
from collections import defaultdict
import os

def extract_student_scores(content):
    """
    从文件内容中提取每个学生的评分数据
    """
    pattern = r'Student (\w+):\s*((?:\d+\.\s*\d+\s*)+|(?:\d+\s*)+)'
    matches = re.findall(pattern, content)
    
    student_scores = defaultdict(list)
    
    for student, scores in matches:
        if '.' in scores:
            score_values = re.findall(r'\d+\.\s*(\d+)', scores)
        else:
            score_values = re.findall(r'\d+', scores)
        
        score_values = [int(value) for value in score_values]
        student_scores[student].append(score_values)
    
    return student_scores

def calculate_average(scores):
    """
    计算评分数据的平均值
    """
    if not scores:
        return 0
    return sum(scores) / len(scores)

def process_files(file_paths):
    """
    处理多个文件，提取每个学生的评分数据并计算平均值
    """
    results = {}
    
    for file_path in file_paths:
        with open(file_path, 'r') as file:
            content = file.read()
        
        student_scores = extract_student_scores(content)
        
        file_results = {}
        for student, scores_list in student_scores.items():
            if len(scores_list) >= 2:
                first_scores = scores_list[0]
                second_scores = scores_list[1]
                
                first_avg = calculate_average(first_scores)
                second_avg = calculate_average(second_scores)
                
                file_results[student] = {
                    'first_scores': first_scores,
                    'first_avg': first_avg,
                    'second_scores': second_scores,
                    'second_avg': second_avg,
                    'difference': abs(first_avg - second_avg)
                }
        
        results[file_path] = file_results
    
    return results

def aggregate_results(results):
    """
    统计所有文档中每位学生的两次数据的平均数和差值
    """
    student_aggregates = defaultdict(lambda: {
        'first_avg_total': 0,
        'second_avg_total': 0,
        'difference_total': 0,
        'count': 0,
        'count_diffs_less_than_05': 0  # 添加用于计数差值小于0.5的次数
    })
    
    # 遍历所有文件的结果
    for file_results in results.values():
        for student, data in file_results.items():
            student_aggregates[student]['first_avg_total'] += data['first_avg']
            student_aggregates[student]['second_avg_total'] += data['second_avg']
            student_aggregates[student]['difference_total'] += data['difference']
            student_aggregates[student]['count'] += 1
            
            # 统计差值小于0.5的次数
            if data['difference'] < 0.5:
                student_aggregates[student]['count_diffs_less_than_05'] += 1
    
    # 计算每位学生的平均值
    for student, aggregates in student_aggregates.items():
        count = aggregates['count']
        aggregates['first_avg_mean'] = aggregates['first_avg_total'] / count
        aggregates['second_avg_mean'] = aggregates['second_avg_total'] / count
        aggregates['difference_mean'] = aggregates['difference_total'] / count
        aggregates['proportion_less_than_05'] = aggregates['count_diffs_less_than_05'] / count  # 计算比例
    
    return student_aggregates

# 获取所有符合条件的文件路径
file_paths = []
for file in os.listdir():
    if file.startswith("consistency_test_20250111") and file.endswith(".txt"):
        file_paths.append(file)

# 处理文件并输出结果
results = process_files(file_paths)

# 统计所有文档中每位学生的两次数据的平均数和差值
student_aggregates = aggregate_results(results)

# 统计文档数
doc_count = len(file_paths)

# 打印统计结果
print(f"文档数量: {doc_count}")
print("所有文档中每位学生的统计结果：")
for student, aggregates in student_aggregates.items():
    print(f"学生: {student}")
    print(f"第一次评分的平均数的平均数: {aggregates['first_avg_mean']:.4f}")
    print(f"第二次评分的平均数的平均数: {aggregates['second_avg_mean']:.4f}")
    print(f"两次评分差值的平均数: {aggregates['difference_mean']:.4f}")
    print(f"差值小于0.5的次数: {aggregates['count_diffs_less_than_05']}")
    print(f"差值小于0.5的比例: {aggregates['proportion_less_than_05']:.2f}")
    print()
```
